## NoRP data 

### load lib

```python
import sys
sys.path.append('/home/ping/project/pyspsw')
from radio.radiolc import  draw
from radio import Norp

datadir='/data/solar/radio'



from scipy.io import readsav
def time2datetime(ntime):
    ##norp time is struct : {milliseconds,mjd}
    ###in astro time the mjd start from :1979-01-01

    ##
    conv=lambda x: datetime(1979,1,1)+timedelta(days=x[1]-1,seconds=x[0]*1.e-3)
    #import pdb 
    #pdb.set_trace()
    norp_time=np.transpose([conv(x.item()) for x in ntime])
    return norp_time

def read_data(datafile,flux_type='fi'):
    data=readsav(datafile)
    #print("Data at "+data.day.decode('ascii')+"!")
    ##data
    fi=np.array(data.fi)
    fiavg=np.array(data.fiavg)
    fv=np.array(data.fv)
    fvavg=np.array(data.fvavg)
    freq=np.array([str('{0:.2f} GHz'.format(tmp)) for tmp in data.freq ])  
    mvd=np.array(data.mvd)
    ntime=np.array(data.tim)

    ##to dataframe
    savdata={'time':time2datetime(ntime),
            'fi':fi,
            'fv':fv,
            'polarization':(fi-fv)/(fi+fv),
            'mvd':mvd,
            'labels':freq
            }
    #print("Averge flux(R+L):",fiavg)
    #print("Averge flux(R-L):",fvavg) 
    norp_df=pd.DataFrame(
            np.multiply(savdata[flux_type],savdata['mvd']),
            index=savdata['time'],
            columns=savdata['labels']
            )
    return  norp_df

#norp_fv=read_data('/data/solar/radio/2013/11/norp20131108_0425.xdr',flux_type='fv')
#norp_fi=read_data('/data/solar/radio/2013/11/norp20131108_0425.xdr',flux_type='fi')
#norp_pol=read_data('/data/solar/radio/2013/11/norp20131108_0425.xdr',flux_type='polarization')

def draw_norp(datafile,st,et,flux_type,ax,subbkg=True):
    norp_df=read_data(datafile,flux_type=flux_type)
    i=0
    for col in norp_df.columns:
        i=i+1
        sindex=np.logical_and(norp_df.index>st+timedelta(seconds=5.0),norp_df.index<=st+timedelta(seconds=35))
        bkg=savgol_filter(norp_df[col][sindex],11,3).mean()#rstndf[col][sindex].min()
        print(bkg)
        if flux_type == 'polarization' :
            bkg=0.0            
        if not subbkg:savgol_filter(norp_df[col]-bkg,19,3)
            bkg=0.0
        ax.plot(norp_df.index,savgol_filter(norp_df[col]-bkg,19,3),alpha=0.8,drawstyle='steps-mid',label=col,lw=1)
        norp_df[col]=savgol_filter(norp_df[col]-bkg,19,3)
    return ax,norp_df

```
